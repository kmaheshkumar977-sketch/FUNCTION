1.What is the difference between a function and a method in Python?
   - Function

     - A function is a block of reusable code that performs a specific task.

     - Defined using the def keyword (or lambda for anonymous functions).

     - It is independent and can exist outside of any class.

     - Called directly by its name.
  - Method

     - A method is also a function, but it is defined inside a class.

     - It operates on objects (instances of the class) and can access/modify their data through self.

     - Called using the dot (.) notation on an object.   

```python
#Example of function
def greet(name):
    return f"Hello, {name}"
print(greet("Mahesh"))  # Calling function
```

```python
#Example of method
class Person:
    def __init__(self, name):
        self.name = name
    def greet(self):
        return f"Hello, {self.name}"
p = Person("Mahesh")
print(p.greet())
```

2. Explain the concept of function arguments and parameters in Python
    - In Python, parameters and arguments are related concepts but they are not the same.

     - A parameter is a variable that is defined inside the parentheses of a function definition. Parameters act like placeholders, telling the function what kind of input it expects. An argument, on the other hand, is the actual value that you pass to the function when you call it. In other words, parameters are like empty boxes defined in the function, and arguments are the real values that get placed inside those boxes when the function runs.

```python
#Example
def greet(name):   # here 'name' is a parameter
    print(f"Hello, {name}!")
greet("Mahesh")    # here "Mahesh" is an argument
```

In this case, name is the parameter, and "Mahesh" is the argument supplied when calling the function.

   - Python also supports different types of arguments, such as:

   - Positional arguments (matched by order),

   - Keyword arguments (matched by name),

   - Default arguments (parameters with default values), and

   - Variable-length arguments (*args and **kwargs for multiple inputs).

So, parameters are defined when creating the function, while arguments are the actual values provided when the function is executed.

5. What are iterators in Python and how do they differ from iterables?
   - Iterables
      - An iterable is any Python object that can return its elements one at a time. Examples include lists, tuples, strings, sets, and dictionaries.

       - They implement the __iter__() method, which returns an iterator.

       - Iterables can be looped over using a for loop.
   - Iterators
     - An iterator is an object that represents a stream of data; it gives you the next element when you call next() on it.

     - It must implement __iter__() (returns itself) and __next__() (returns the next value).

     - When no items are left, __next__() raises a StopIteration exception.    

```python
#Example of Iterables
my_list = [1, 2, 3]
for item in my_list:   # list is an iterable
    print(item)
```

```python
#Example of lterators
my_list = [1, 2, 3]
it = iter(my_list)  # converting iterable into an iterator
print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
# next(it) now will raise StopIteration
```

6.Explain the concept of generators in Python and how they are defined
  - In Python, a generator is a special type of iterator that allows you to generate values on the fly, rather than storing them all in memory at once. This makes generators very memory efficient and useful when dealing with large datasets or infinite sequences.

  - Generators are defined using a function with the yield keyword instead of return. When the function is called, it doesn’t execute the entire code immediately. Instead, it returns a generator object, which produces one value at a time whenever next() is called or when used in a loop.

```python
def count_up_to(n):
    num = 1
    while num <= n:
        yield num   # yield pauses the function and returns a value
        num += 1

# Using the generator
gen = count_up_to(5)
for value in gen:
    print(value)
```

7.What are the advantages of using generators over regular functions?
   - Memory Efficiency

     - Regular functions that return lists (or other collections) store the entire result in memory at once.

     - Generators yield items one by one, so they don’t need to keep the whole sequence in memory.

     - This makes them ideal for processing large data streams or infinite sequences.
   - Lazy Evaluation (On-Demand Execution)

     - Generators compute values only when needed, instead of precomputing everything.

     - This saves time and resources when you don’t need all results at once.
   - Infinite Sequences

     - Regular functions cannot easily handle infinite sequences because they would try to build an infinite list.

     - Generators can produce values indefinitely, since they yield results one at a time.   

8.What is a lambda function in Python and when is it typically used?
   - A lambda function in Python is a small, anonymous function defined using the lambda keyword instead of the usual def. Unlike regular functions, a lambda function can have any number of arguments but only a single expression, which is automatically returned. Its general form is:
     - lambda arguments: expression
   - Lambda functions are typically used when you need a short, one-time function and don’t want to formally define it using def. Common use cases include:

     - As a key function in sorting (sorted, min, max)

     - In functional programming tools like map(), filter(), and reduce()

     - For quick, throwaway operations in places where defining a full function would be unnecessary.  

```python
#Example
square = lambda x: x * x
print(square(5))  # Output: 25


```

9.Explain the purpose and usage of the map() function in Python
   - The map() function in Python is used to apply a function to each item in an iterable (like a list, tuple, or string) and returns a map object (an iterator) containing the results.
   - Purpose of map()
     - To perform an operation on every element of an iterable without writing an explicit loop.

     - It helps make the code more concise and readable compared to using for loops.

     - It supports functional programming by allowing functions to be applied directly to data.
     - map() returns an iterator, so you usually convert it into a list or tuple if you want to see the results immediately.

     - It’s often used with lambda functions or built-in functions.

     - More memory efficient than list comprehensions if you only need to iterate once (since it produces results lazily).
  - When to use:
     - Use map() when you want to apply a transformation function to every element of an iterable, especially if the function already exists (like str.upper, abs, etc.) or when you want cleaner code instead of writing loops.   

```python
#Example
# Convert numbers to their squares
numbers = [1, 2, 3, 4, 5]
squares = map(lambda x: x**2, numbers)
print(list(squares))
```

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
   - Map()
      - Purpose: Apply a function to every item of an iterable.

      - Input: function + iterable(s)

      - Output: new iterable (with the function applied to each element)
 - filter()
      - Purpose: Filter elements of an iterable based on a condition (function returning True or False).

     - Input: predicate function (boolean) + iterable

     - Output: new iterable (only elements where function returned True)
  - reduce()
    - Purpose: Reduce an iterable to a single cumulative value by applying a function repeatedly.
     (Unlike map() and filter(), reduce() is in the functools module.)

    - Input: function (with 2 arguments) + iterable

    - Output: single value    

```python
#Map Example
numbers = [1, 2, 3, 4, 5]
squares = map(lambda x: x**2, numbers)
print(list(squares))   # [1, 4, 9, 16, 25]
```

```python
#Filter Example
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))   # [2, 4, 6]
```

```python
#Reduce Example
from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)   # 120

```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:   # check if even
            total += num
    return total


my_list = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(my_list))   # Output: 12

```

```python
#4.Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False   # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms   # total terms to generate
        self.count = 0               # how many terms generated so far
        self.a, self.b = 0, 1        # starting values for Fibonacci

    def __iter__(self):
        return self   # iterator must return itself

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration   # stop iteration when limit reached

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")


```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in power_of_two(5):
    print(power)
```

```python
#9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Function to convert Celsius to Fahrenheit
def c_to_f(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to apply conversion
fahrenheit_temps = list(map(c_to_f, celsius_temps))

# Display results
print("Celsius:    ", celsius_temps)
print("Fahrenheit: ", fahrenheit_temps)

```

```python
#10.Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    # filter out characters that are NOT vowels
    result = filter(lambda ch: ch not in vowels, text)
    return "".join(result)

# Example usage
input_str = "Hello World, Python is fun!"
output_str = remove_vowels(input_str)

print("Original String:", input_str)
print("Without Vowels: ", output_str)
```
